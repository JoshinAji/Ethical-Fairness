```python
import pandas as pd
import numpy as np

v1 = pd.read_csv("data/labels_v1.csv")

# Decide balance target per split (avoid leakage)
def balance(df, keys=['race_cat']):
    # Drop rows with missing race_cat
    df = df.dropna(subset=keys)
    # Find target count = min class size (cap at reasonable K)
    counts = df.groupby(keys).size()
    K = counts.min()
    balanced = (df.groupby(keys, group_keys=False)
                  .apply(lambda g: g.sample(min(len(g), K), random_state=42))
                  .reset_index(drop=True))
    return balanced

v2_list = []
for split in ['train','val','test']:
    part = v1[v1['split']==split].copy()
    # Option A: balance on race only
    b = balance(part, ['race_cat'])
    # Option B (comment in if needed): balance on race√ógender
    # b = balance(part.dropna(subset=['gender_cat']), ['race_cat','gender_cat'])
    v2_list.append(b)

v2 = pd.concat(v2_list, ignore_index=True)
v2.to_csv("data/labels_v2_balanced.csv", index=False)
```
